Cargue los securities indicados desde **2025-01-01 hasta hoy** en LocalDb.

```python
# Creación de TABLA HISTORICA y Vista para Reporte
```

```python
def initialize_database():
    global db, logger, logging
    try:
        with open('vmetrix/database.sql', 'r') as f:
            sql_script = f.read()
        db.command(sql_script)
        logger.info("Tablas creadas.")
    except Exception as e:
        logger.info(f"Error DB: {e}")

initialize_database()
```

```python
# Carga de Tabla HISTORICA
```

```python
import os
import pandas as pd
from datetime import date
from config import SECURITY_LIST, SECURITY_LIST_STR, PATH_RAW, PATH_INC

def run_historical_load():
    global db, logger, logging, banxico
    logging.getLogger("vmetrix").disabled = True
    logger.info(f"Iniciando Carga Histórica para los activos: {SECURITY_LIST}")
    os.makedirs(PATH_RAW, exist_ok=True)
    
    # 1. Obtener metadatos de BANXICO_SERIES
    query_filt = f"SELECT IDSERIE, SECURITY_NAME FROM BANXICO_SERIES WHERE SECURITY_NAME IN ({SECURITY_LIST_STR})"
    securities_df = db.query(query_filt)
    
    START_DATE = "2025-01-01"
    END_DATE = date.today().strftime("%Y-%m-%d")

    for _, row in securities_df.iterrows():
        series_id = row['IDSERIE']
        security_name = row['SECURITY_NAME']
        
        try:
            logger.info(f"Procesando {security_name} ({series_id})...")
            
            # 2. Extracción desde Banxico
            response = banxico.get_values_between(series_id, START_DATE, END_DATE)

            logging.getLogger("vmetrix").disabled = False
            
            # Navegar la estructura
            try:
                data_list = response['bmx']['series'][0]['datos']
            except (KeyError, IndexError):
                logger.info(f"   Estructura inesperada o sin datos para {series_id}")
                continue

            # 3. Transformación y Enriquecimiento
            # Elección: Python (Pandas). Se utiliza por su flexibilidad para manipular estructuras 
                
            df = pd.DataFrame(data_list)
            df = df.rename(columns={'fecha': 'date', 'dato': 'value'})
            
            df['series_id'] = series_id
            df['security_name'] = security_name
            df['updated_at'] = pd.Timestamp.now()
            
            # Limpieza de tipos y orden de columnas
            df['date'] = pd.to_datetime(df['date'], dayfirst=True).dt.date
            df['value'] = pd.to_numeric(df['value'], errors='coerce')
            df = df[['date', 'series_id', 'security_name', 'value', 'updated_at']]
            
            # 4. Capa RAW (Parquet)
            # Elección: Python (PyArrow). Se utiliza para garantizar la persistencia de la primera cara. 

            # path_archivo = os.path.join(PATH_RAW, f"raw_hist_{series_id}.parquet")                        
            run_date = date.today().strftime("%Y-%m-%d")
            path_dir = os.path.join(PATH_RAW, security_name, run_date)
            os.makedirs(path_dir, exist_ok=True)
            path_archivo = os.path.join(path_dir, "raw.parquet")
            
            df.to_parquet(path_archivo, index=False)
            
            # 5. Carga Idempotente a DuckDB
            # El borrado se ejecutan en el motor de BD para garantizar cargar siempre lo mismo.
            
            db.command(f"DELETE FROM fact_securities WHERE series_id = '{series_id}' AND date >= '{START_DATE}'")
            db.write_df("fact_securities", df)
            
            logger.info(f"   Éxito: {len(df)} registros insertados.")
            
        except Exception as e:
            logger.info(f"   Error en {series_id}: {e}")

# Ejecutar carga histórica filtrada
run_historical_load()
```

### 2. Carga diaria incremental

Implemente la arquitectura descrita en la imagen anterior para que el pipeline pueda correr todos los días como un proceso diario.

**2.1 Valores del día**
- Traer los valores vigentes al día de hoy para los cuatro securities.
- Para `UDIMXN.SPOT` ver el caso especial abajo (**NO** se asume un solo valor diario).

**2.2 Estadísticas móviles (7 días)**
Cree una tabla de estadísticas de los últimos **7 días calendario** (ventana `[value_date - 6, value_date]`)
- Evalúe si es necesario una tabla, vista o función para resolver este problema.
    - Justifique su respuesta.

Se usan **7 días calendario** (no hábiles) para que la definición sea la misma para todos los securities

```python
import os
import pandas as pd
from datetime import date
from config import SECURITY_LIST, SECURITY_LIST_STR, PATH_INC

def run_daily_incremental():
    global db, logger, logging, banxico
    logging.getLogger("vmetrix").disabled = True
    """
    Realiza la carga incremental diaria de los activos definidos en SECURITY_LIST.
    Sigue el flujo: API -> DataFrame -> Parquet (inc/) -> DuckDB.
    """
    logger.info(f"Iniciando Carga Diaria Incremental para: {SECURITY_LIST}")
    
    # Asegurar que la subcarpeta s3_datalake/inc/ exista
    os.makedirs(PATH_INC, exist_ok=True)
    
    # 1. Obtener metadatos desde la base de datos local
    # DuckDB. Se prefiere SQL para el filtrado de metadatos de configuración por ser mas eficiente.
    query_filt = f"""
        SELECT IDSERIE, SECURITY_NAME 
        FROM BANXICO_SERIES 
        WHERE SECURITY_NAME IN ({SECURITY_LIST_STR})
    """
    securities_df = db.query(query_filt)

    if securities_df.empty:
        logger.info("No se encontraron metadatos para los activos configurados.")
        return

    for _, row in securities_df.iterrows():
        series_id = row['IDSERIE']
        security_name = row['SECURITY_NAME']
        
        try:
            
            # 2. Extracción desde la API de Banxico (Solo el último punto)
            response = banxico.get_last_value(series_id)
            
            try:
                data_list = response['bmx']['series'][0]['datos']
            except (KeyError, IndexError):
                logger.info(f"  Banxico no retornó datos nuevos para {series_id} hoy.")
                continue

            # 3. Transformación y Enriquecimiento
            # Python (Pandas). Se utiliza para el formateo inicial de la respuesta del API 
            # porque Pandas es más eficiente para transformar estructuras JSON/Listas anidadas en tablas 
            df = pd.DataFrame(data_list)
            df = df.rename(columns={'fecha': 'date', 'dato': 'value'})
            
            # Formateo de tipos y metadatos
            df['series_id'] = series_id
            df['security_name'] = security_name
            df['updated_at'] = pd.Timestamp.now()
            df['date'] = pd.to_datetime(df['date'], dayfirst=True).dt.date
            df['value'] = pd.to_numeric(df['value'], errors='coerce')
            
            # Orden de columnas final (esquema de la tabla fact_securities)
            df = df[['date', 'series_id', 'security_name', 'value', 'updated_at']]

            # Caso especial UDIMXN.SPOT: Banxico publica valores futuros en bloques.
            if security_name == 'UDIMXN.SPOT':
                today = date.today()
                df = df[df['date'] <= today]
                logger.info(f"  UDIMXN.SPOT: {len(df)} registros vigentes al {today} (se descartaron fechas futuras)")
                
                if df.empty:
                    logger.info(f"  UDIMXN.SPOT: sin valores nuevos para hoy")
                    continue            
            
            # 4. Persistencia en Capa Incremental (Parquet)
            # today_str = date.today().strftime("%Y%m%d")
            # file_name = f"raw_inc_{series_id}_{today_str}.parquet"
            # path_full = os.path.join(PATH_INC, file_name)

            run_date = date.today().strftime("%Y-%m-%d")
            path_dir = os.path.join(PATH_INC, security_name, run_date)
            os.makedirs(path_dir, exist_ok=True)
            path_full = os.path.join(path_dir, "raw.parquet")
            
            df.to_parquet(path_full, index=False)
            
            # 5. Carga Idempotente en DuckDB
            # El borrado se ejecutan en el motor de BD para garantizar cargar siempre lo mismo.
            
            current_date = df['date'].iloc[0]
                        
            db.command(f"""
                DELETE FROM fact_securities 
                WHERE series_id = '{series_id}' 
                AND date = '{current_date}'
            """)
            
            # Insertamos el nuevo registro
            db.write_df("fact_securities", df)
            
            logger.info(f" Éxito: Registro {current_date} guardado en {path_full} e insertado en DB.")
            
        except Exception as e:
            logger.info(f"  Error procesando {security_name}: {e}")

# Ejecución del proceso
run_daily_incremental()
```

```python
import os
import pandas as pd
from datetime import datetime
from config import PATH_REPORT

def run_quality_report():
    global db, logger, logging, banxico
    logging.getLogger("vmetrix").disabled = True
    
    os.makedirs(PATH_REPORT, exist_ok=True)
    
    query = """
    SELECT 
        security_name,
        MIN(date) as fecha_inicio,
        MAX(date) as fecha_fin,
        COUNT(*) as total_registros,
        ROUND(AVG(value), 4) as promedio,
        MAX(value) as maximo,
        MIN(value) as minimo,
        COUNT(*) FILTER (WHERE value IS NULL) as nulos,
        MAX(updated_at) as ultima_actualizacion_dato
    FROM fact_securities
    GROUP BY security_name
    ORDER BY security_name
    """
    
    report_df = db.query(query)
    
    timestamp_str = datetime.now().strftime("%Y%m%d%H%M%S")
    report_filename = f"reporte_validacion_{timestamp_str}.csv"
    path_full = os.path.join(PATH_REPORT, report_filename)
    
    # 1. Exportar reporte
    report_df.to_csv(path_full, index=False)
    
    logger.info("Reporte generado exitosamente.")
    logger.info(f"Ubicación: {path_full}")

run_quality_report()
```

### Caso especial `UDIMXN.SPOT`

UDIS **sí** tiene un valor **diario** — no son dos datos al mes. Lo que ocurre es que Banxico publica el valor de UDIS **con anticipación en bloques**:

- Alrededor del **día 10** de cada mes se publica el horizonte de valores diarios hasta el **día 25** siguiente.
- Alrededor del **día 25** de cada mes se publica el horizonte hasta el **día 10** del mes siguiente.

Consecuencias prácticas para el pipeline:
1. `get_last_value('SP68257')` devuelve la fecha del **horizonte publicado** (puede estar varios días en el futuro respecto al runtime), *NO* el valor de hoy.
