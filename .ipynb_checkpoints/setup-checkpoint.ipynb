```python
import logging
import os
from datetime import datetime

# --- 1. Configuración de nombres y carpetas ---
timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
log_filename = f"logs/pipeline_{timestamp}.log"
os.makedirs("logs", exist_ok=True)

logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    handlers=[
        logging.FileHandler(log_filename, encoding='utf-8'),
        logging.StreamHandler()
    ],
    force=True 
)

logger = logging.getLogger("PIPELINE")

# --- 2. SILENCIAR LOGS ---
loggers_a_silenciar = [
    'vmetrix', 
    'vmetrix.banxico_api', 
    'vmetrix.database', 
    'httpx', 
    'httpcore'
]

for log_name in loggers_a_silenciar:
    l = logging.getLogger(log_name)
    l.setLevel(logging.WARNING) # Ignora los INFO, solo muestra Errores
    l.propagate = False         # Corta la comunicación con tu consola

logger.info("Sistema de logging inicializado.")
```
