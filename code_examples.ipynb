#### 1. Banxico REST API Wrapper
El package ``vmetrix`` provee un cliente HTTP para la API REST de Banxico (Banco Central de México).

> El token de acceso a la API se encuentra en el archivo ``.env`` del proyecto.

```python
from vmetrix.config import get_config

# Ver token Banxico API
CONFIG = get_config()
CONFIG.TOKEN_BANXICO
```

```python
from vmetrix import get_banxico_api

# instanciar wrapper
api = get_banxico_api()
api
```

##### 1.1 Banxico Series
Corresponde a los **identificadores** de los datos de mercado que pueden ser extraidos de la API. Por ejemplo, el valor del tipo de cambio FIX (observado) para el dólar de USA versus el peso mexicano se puede obtener con la serie `SF43718`.
    
* El nombre corto de esta serie sería el SECURITY_NAME, en este caso: `USDMXN.FIX`

```python
series = "SF43718,SP68257"

# Consulta al endpoint de metadatos para un par de series
api.get_metadata(series=series)
```

```python
# Consulta al endpoint para obtener los ultimos datos de un listado de series
api.get_last_value(series=series)
```

```python
# Consulta de datos para un listado de series entre dos fechas
api.get_values_between(
    series=series,
    start_date="2026-03-01",
    end_date="2026-12-31"
)
```

```python
from vmetrix import get_database

# Para obtener las series y su serie, se puede consultar la base de datos local, especificamente la tabla BANXICO_SERIES
db = get_database()

sql = """
SELECT *
FROM BANXICO_SERIES
WHERE IDSERIE = $I_IDSERIE
"""

db.query(sql, params={"I_IDSERIE": "SF43718"})
```
